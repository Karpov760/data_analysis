Постройте самостоятельно гистограмму распределения стоимости билета среди выживших и умерших на одном графике

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns


df = pd.read_csv('titanic.csv')
fig, ax = plt.subplots()

df_s1 = df[(df.Survived == 1)]
df_s0 = df[(df.Survived == 0)]

d1_heights, d1_bins = np.histogram(df_s1['Fare'], bins = 8)
d0_heights, d0_bins = np.histogram(df_s0['Fare'], bins=d1_bins)

width = (d1_bins[1] - d1_bins[0])/3

ax.bar(d1_bins[:-1], d1_heights, width=width, facecolor='cornflowerblue', label = 'Survived')
ax.bar(d0_bins[:-1]+width, d0_heights, width=width, facecolor='seagreen', label = 'Not survived')
plt.ylabel('Fare')
plt.legend()
```

Т.е. чем ниже цена билета, тем больше шансов умереть(((
